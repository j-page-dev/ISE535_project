```python
'''
This script downloads the offensive statistics and stores them in a offense_db.csv file
'''
# import packages
# pip install sportsipy
from sportsipy.nfl.teams import Teams
import pandas as pd

# read in the data for each of the 32 NFL teams
teams = Teams()

# split out into all 32 teams
packers = teams('GNB')
bills = teams('BUF')
bucs = teams('TAM')
titans = teams('OTI')
saints = teams('NOR')
chiefs = teams('KAN')
ravens = teams('RAV')
seahawks = teams('SEA')
colts = teams('CLT')
raiders = teams('RAI')
vikings = teams('MIN')
steelers = teams('PIT')
cardinals = teams('CRD')
browns = teams('CLE')
dolphins = teams('MIA')
falcons = teams('ATL')
cowboys = teams('DAL')
texans = teams('HTX')
chargers = teams('SDG')
lions = teams('DET')
niners = teams('SFO')
rams = teams('RAM')
bears = teams('CHI')
panthers = teams('CAR')
redskins = teams('WAS')
eagles = teams('PHI')
patriots = teams('NWE')
broncos = teams('DEN')
bengals = teams('CIN')
jaguars = teams('JAX')
giants = teams('NYG')
jets = teams('NYJ')

# extract the list of players on each team roster
packers_roster = packers.roster.players
bills_roster = bills.roster.players
bucs_roster = bucs.roster.players
titans_roster = titans.roster.players
saints_roster = saints.roster.players
chiefs_roster = chiefs.roster.players
ravens_roster = ravens.roster.players
seahawks_roster = seahawks.roster.players
colts_roster = colts.roster.players
raiders_roster = raiders.roster.players
vikings_roster = vikings.roster.players
steelers_roster = steelers.roster.players
cardinals_roster = cardinals.roster.players
browns_roster = browns.roster.players
dolphins_roster = dolphins.roster.players
falcons_roster = falcons.roster.players
cowboys_roster = cowboys.roster.players
texans_roster = texans.roster.players
chargers_roster = chargers.roster.players
lions_roster=lions.roster.players
niners_roster = niners.roster.players
rams_roster = rams.roster.players
bears_roster = bears.roster.players
panthers_roster = panthers.roster.players
redskins_roster = redskins.roster.players
eagles_roster = eagles.roster.players
patriots_roster = patriots.roster.players
broncos_roster = broncos.roster.players
bengals_roster = bengals.roster.players
jaguars_roster = jaguars.roster.players
giants_roster = giants.roster.players
jets_roster = jets.roster.players

# download previous year data for the players on the active rosters from the prior steps
seasons=['2020'] # this is very slow to download for multiple seasons!
df = pd.DataFrame(columns = ['Date', 'Team', 'Opp', 'Name', 'Pos', 'FD_points'])
all_players = [packers_roster,bills_roster,bucs_roster, titans_roster, saints_roster, chiefs_roster, ravens_roster, seahawks_roster, colts_roster, raiders_roster, vikings_roster, steelers_roster, cardinals_roster, browns_roster, dolphins_roster,falcons_roster, cowboys_roster, texans_roster, chargers_roster, lions_roster,niners_roster, rams_roster, bears_roster, panthers_roster, redskins_roster, eagles_roster, patriots_roster, broncos_roster, bengals_roster, jaguars_roster, giants_roster, jets_roster]

# web scrape the individual player data for each active player in each of the specified seasons
for k in seasons:
    for i in all_players:    
        name=[]    
        for j in range(len(i)):          
            temp = pd.DataFrame(columns = ['Date', 'Team', 'Opp', 'Name','Key', 'Pos', 'FD_points'])
            name.append(repr(i[j]))
            player = name[j]
            key = player[player.find('(')+1:player.find(')')]
            first_letter = key[0]
            url = 'https://www.pro-football-reference.com/players/' + first_letter + '/' + key + '/fantasy/' + k + '/'
            data = pd.read_html(url)[0]
            temp.Date = (data.iloc[:,2]) # date of matchup
            temp.Team = (data.iloc[:,3]) # team of player
            temp.Opp = (data.iloc[:,5]) # opposing team
            temp.Name = player # player name
            temp.Key = key # player unique key
            temp.Pos = (data.iloc[:,7]) # player position
            temp.FD_points = (data.iloc[:,-1]) # player FanDuel Fantasy Points for the matchup
            df = df.append(temp)   
            df.dropna(inplace=True) # drop games with no statistics for the player (i.e. they were inactive but on roster)
            
# save the dataset to a csv file
df.to_csv('offense_db.csv')
```
